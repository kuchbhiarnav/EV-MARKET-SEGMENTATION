# Kmeans

```python
features = list(A.columns)

from sklearn.preprocessing import LabelEncoder

cols = ['Profession','Marrital Status','Education','Car_Loan']
A[cols] = A[cols].apply(LabelEncoder().fit_transform)
A.head()
cluster_data = A[features]
```

```python
A.head()
```

```python
from sklearn.cluster import KMeans
clustering_kmeans = KMeans(n_clusters=2)
cluster_data['clusters'] = clustering_kmeans.fit_predict(cluster_data)
cluster_data['clusters'].value_counts()
```

```python
Cluster_0 = cluster_data[cluster_data.clusters==0]
Cluster_1 = cluster_data[cluster_data.clusters==1]

plt.scatter(Cluster_0.Price, Cluster_0['Total Salary'],color='red')
plt.scatter(Cluster_1.Price, Cluster_1['Total Salary'],color='green')

plt.xlabel('Price')
plt.ylabel('Total salary')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

x = cluster_data['Age'].values.reshape(-1, 1)
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
cluster_data['Age']= x_scaled

x = cluster_data['Total Salary'].values.reshape(-1, 1)
x_scaled = scaler.fit_transform(x)
cluster_data['Total Salary']= x_scaled

x = cluster_data['Price'].values.reshape(-1, 1)
x_scaled = scaler.fit_transform(x)
cluster_data['Price']= x_scaled
```

```python
cluster_data['clusters'] = clustering_kmeans.fit_predict(cluster_data) 
cluster_data
```

```python
Cluster_0 = cluster_data[cluster_data.clusters==0]
Cluster_1 = cluster_data[cluster_data.clusters==1]
Cluster_2 = cluster_data[cluster_data.clusters==2]
plt.scatter(Cluster_0.Price, Cluster_0['Total Salary'],color='red')
plt.scatter(Cluster_1.Price, Cluster_1['Total Salary'],color='green')
plt.scatter(Cluster_2.Price, Cluster_2['Total Salary'],color='yellow')

plt.xlabel('Price')
plt.ylabel('Total salary')
plt.show()
```

```python
clustering_kmeans = KMeans(n_clusters=3)
cluster_data['clusters'] = clustering_kmeans.fit_predict(cluster_data) 
```

```python
cluster_data['clusters'].value_counts()
```

```python
clustering_kmeans.cluster_centers_
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(cluster_data)
    sse.append(km.inertia_)
```

```python
plt.title('The Elbow Point Graph')
plt.xlabel('K')
plt.ylabel('sse')
plt.plot(k_range,sse)
```

```python
labels = clustering_kmeans.labels_
labels
```

```python
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(111, projection='3d')
ax.scatter(cluster_data.Age[labels == 0], cluster_data["Total Salary"][labels == 0], cluster_data["Price"][labels == 0],  c='darkblue', s=60)
ax.scatter(cluster_data.Age[labels == 1], cluster_data["Total Salary"][labels == 1], cluster_data["Price"][labels == 1],  c='red', s=60)
ax.scatter(cluster_data.Age[labels == 2], cluster_data["Total Salary"][labels == 2], cluster_data["Price"][labels == 2],  c='green', s=60)
ax.view_init(30, 185)

plt.xlabel("Age")
plt.ylabel("Total Salary")
ax.set_zlabel('Price')
plt.show()
```
